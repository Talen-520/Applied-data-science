````python


# Step 1: Cluster the training dataset using KMeans

k_values = range(2, 11)
kmeans_models = [KMeans(n_clusters=k, random_state=23655842).fit(df_train) for k in k_values]


# Step 3: Compute mean inertia

# Here's the total_inertia function provided by the prompt:
def total_inertia(data, model):
    """Compute the inertia for data given a KMeans model."""
    distance = model.transform(data)
    closest = np.argmin(distance, axis=1)
    inertia = sum([distance[i, closest[i]]**2 for i in range(len(closest))])
    return inertia

train_inertia = [model.inertia_ for model in kmeans_models]
valid_inertia = [total_inertia(df_valid, model) for model in kmeans_models]

# Step 4: Plot the mean inertias
plt.figure(figsize=(12, 6))
plt.plot(k_values, train_inertia, '-o', label="Training Data")
plt.plot(k_values, valid_inertia, '-s', label="Validation Data")
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Mean Inertia')
plt.title('Elbow Curve for Optimal k')
plt.legend()
plt.show()

````

````python
# Step 5: Compute and plot silhouette scores
silhouette_scores = [silhouette_score(df_valid, model.predict(df_valid)) for model in kmeans_models]

plt.figure(figsize=(12, 6))
plt.plot(k_values, silhouette_scores, '-o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k values')
plt.show()
````

````python

def r2_score_tf(y_true, y_pred):
    residual = tf.reduce_sum(tf.square(tf.subtract(y_true, y_pred)))
    total = tf.reduce_sum(tf.square(tf.subtract(y_true, tf.reduce_mean(y_true))))
    r2_score = tf.subtract(1.0, tf.divide(residual, total))
    return r2_score
````
